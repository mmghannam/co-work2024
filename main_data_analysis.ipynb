```python
import numpy as np
import matplotlib.pyplot as plt

instance_sizes = [(len(instance['couriers']), len(instance['deliveries']), instance['instance_name']) for instance in all_instances]
#print(instance_sizes)

sorted_instance_sizes = sorted(instance_sizes, key=lambda x: x[0] / [1])
print(sorted_instance_sizes)

plt.plot([x[0] for x in sorted_instance_sizes], [x[1] for x in sorted_instance_sizes], 'o', markersize=3)

# Calculate the regression line
regression_line = np.polyfit([x[0] for x in sorted_instance_sizes], [x[1] for x in sorted_instance_sizes], 1)
x_values = [x[0] for x in sorted_instance_sizes]
y_values = np.polyval(regression_line, x_values)

# Plot the regression line
plt.plot(x_values, y_values, 'r-', label='Linear Regression')

# Linear regression equation
slope = regression_line[0]
intercept = regression_line[1]
equation = f'y = {slope:.2f}x + {intercept:.2f}'
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes)

# Rest of the code
plt.xlabel('Number of couriers')
plt.ylabel('Number of deliveries')
plt.title('Instance sizes')
plt.legend()
plt.show()
```
